```python
import numpy as np
import pandas as pd
data = pd.read_csv('Documents/columbia/APAN005/PAC PROJECT/analysis_data.csv')
```

```python
#Remove NaNs with function below
```

```python
data = data.fillna(0) # replace NaNs to 0
```

```python
#Confirm NaNs were replaced to 0
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
y = data.monthly_spend
X = data.drop('monthly_spend', axis=1)
```

```python
X
```

```python
#Convert Data Types 'Object' to 'Values'
```

```python
categorical_features = ['gender','marital_status','education_level','region','employment_status','card_type']

X = pd.get_dummies(X, 
                         prefix_sep = '_', 
                         columns = categorical_features, 
                         drop_first = True)

X.reindex(columns=X.columns, fill_value=0)
X.head()
```

```python
len(X.columns)
```

```python
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.metrics import root_mean_squared_error
tree = DecisionTreeRegressor(random_state=617, 
                             max_depth = 20)  # Default is a maximal tree
tree.fit(X,y)
```

```python
pred_train = tree.predict(X)
```

```python
pred_train
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(14, 10))
plot_tree(tree, 
          feature_names=X.columns,
          label = 'root', 
          filled=True,
          fontsize=8,
          rounded = True, 
          precision = 1,
          node_ids=False)
plt.title("Regression Tree")
plt.show()
```

```python
from sklearn.metrics import root_mean_squared_error
train_pred = tree.predict(X)
rmse = root_mean_squared_error(y, train_pred)
rmse
```
