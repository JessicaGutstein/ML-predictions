```python
import pandas as pd
data = pd.read_csv('Documents/columbia/APAN005/PAC PROJECT/analysis_data.csv')
```

```python
y = data.monthly_spend
X = data.drop('monthly_spend', axis=1)
```

```python
data.columns
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X.loc[:,['credit_score', 'credit_limit','annual_income','avg_transaction_value']], y)
```

```python
from sklearn.metrics import root_mean_squared_error
pred = reg.predict(X.loc[:,['credit_score', 'credit_limit','annual_income','avg_transaction_value']])
rmse = root_mean_squared_error(y, pred)
rmse
```

```python
# Test the model on new unseen data.
scoring_data = pd.read_csv('Documents/columbia/APAN005/PAC PROJECT/scoring_data.csv')
scoring_data_X = scoring_data.loc[:,['credit_score', 'credit_limit','annual_income','avg_transaction_value']]
pred = reg.predict(scoring_data_X)
```
