```python
import numpy as np
import pandas as pd
data = pd.read_csv('Documents/columbia/APAN005/PAC PROJECT/analysis_data.csv')
```

```python
data = data.fillna(0) # replace NaNs to 0
```

```python
#Confirm NaNs were replaced to 0
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
y = data.monthly_spend
X = data.drop('monthly_spend', axis=1)
```

```python
X
```

```python
#Convert Data Types 'Object' to 'Values'
```

```python
categorical_features = ['gender','marital_status','education_level','region','employment_status','card_type']

X = pd.get_dummies(X, 
                         prefix_sep = '_', 
                         columns = categorical_features, 
                         drop_first = True)

X.reindex(columns=X.columns, fill_value=0)
X.head()
```

```python
len(X.columns)
```

```python
#PCA
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=29)
pca.fit(X, y)
X_train_pc = pca.transform(X)
X_train_pc
```

```python
X_train_pc.shape
```

```python
import numpy as np
np.sum(pca.explained_variance_ratio_)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train_pc, y)
pred_train = reg.predict(X_train_pc)
```

```python
pd.DataFrame(X_train_pc)
```

```python
from sklearn.metrics import root_mean_squared_error
train_pred = reg.predict(X_train_pc)
rmse = root_mean_squared_error(y, train_pred)
rmse
```
